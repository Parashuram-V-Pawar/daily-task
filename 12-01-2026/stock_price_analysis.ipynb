```python
# Problem Statement: 
# 2) A DataFrame contains date, stock_name, open_price, and close_price.
# -> Task:
#   1. Calculate percentage daily return
#   2. Identify top gaining and losing stocks per month
#   3. Compute rolling volatility
```

```python
import pandas as pd
```

```python
stock_data = pd.read_csv("company_data.csv")
stock_data
```

```python
# Calculate percentage daily return
stock_data['Daily Return'] = (stock_data['close_price'] - stock_data['open_price'])/ stock_data['open_price'] * 100
stock_data
```

```python
# Identify top gaining and losing stocks per month
stock_group = stock_data.groupby("stock_name")
return_sum = stock_group["Daily Return"].sum()
for x in return_sum.index:
    if return_sum[x] > 0:
        print(f"{x} is a Gaining Stock with total return of {return_sum[x]:.2f}%")
    else:
        print(f"{x} is a Losing Stock with total return of {return_sum[x]:.2f}%")
```

```python
# Compute rolling volatility on return_sum
rolling_volatility = return_sum.rolling(window=3).std()
rolling_volatility
```
