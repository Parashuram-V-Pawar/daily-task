```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/sales_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.cloumns =df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

df['Month'] = df['Date'].dt.month

df = df.drop(columns=['Date'])
```

```python
df.head()
```

```python
from numpy._core import numeric
X = df.drop(columns=['Total Sales'])
y = df['Total Sales']

categorical_columns = ['Salesperson','Product','Region']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns)
    ],
    remainder = 'passthrough'
    )
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
pipeline = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('model', RandomForestRegressor())
]
)
```

```python
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
```

```python
r_score = r2_score(y_test, y_pred)
print("R2 Score : ",r_score)

mean_absoltute = mean_absolute_error(y_test, y_pred)
print("Mean absolute error: ",mean_absoltute)

mean_squared = mean_squared_error(y_test, y_pred)
print("Mean squared error: ",mean_squared)
```

```python
test_data = {
  "Salesperson": "Alice",
  "Product" : "Laptop",
  "Region" : "South",
  "Item Unit" : 1300,
  "No. of Items" : 10,
  "Month": 10
}
predicted_sales = pipeline.predict(pd.DataFrame([test_data]))
print("Predicted sales: ",predicted_sales)
```
